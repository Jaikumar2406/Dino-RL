## 1. What was the average active power consumption in March 2007?

```python
march_2007 = df.loc['2007-03']
avg_power_march = march_2007['Global_active_power'].mean()
print(f"Average active power consumption in March 2007: {avg_power_march:.2f} kW")
```

```python
march_2007 = df.loc['2007-03']
daily_power_march = march_2007['Global_active_power'].resample('D').sum()

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(daily_power_march.index, daily_power_march, marker='o', linestyle='-', color='b', label='Daily Power Consumption (kWh)')
plt.title('Active Power Consumption in March 2007')
plt.xlabel('Date')
plt.ylabel('Global Active Power (kWh)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 2. What hour of the day had the highest power usage on Christmas 2006?

```python
christmas_2006 = df.loc['2006-12-25']
hourly_power = christmas_2006.groupby(christmas_2006.index.hour)['Global_active_power'].mean()
max_hour = hourly_power.idxmax()

plt.figure(figsize=(10, 5))
hourly_power.plot(kind='bar')
plt.title('Average Power Consumption by Hour on Christmas 2006')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Global Active Power (kW)')
plt.axvline(x=max_hour, color='r', linestyle='--', label=f'Peak Hour: {max_hour}:00')
plt.legend()
plt.show()

print(f"The hour with highest power usage on Christmas 2006 was: {max_hour}:00")
```

## 	3. Compare energy usage (Global_active_power) on weekdays vs weekends.

```python
df['day_of_week'] = df.index.dayofweek
df['is_weekend'] = df['day_of_week'] >= 5

weekend_avg = df[df['is_weekend']]['Global_active_power'].mean()
weekday_avg = df[~df['is_weekend']]['Global_active_power'].mean()

print(f"Weekday Avg: {weekday_avg:.2f} kW, Weekend Avg: {weekend_avg:.2f} kW")
```

```python
plt.figure(figsize=(8, 6))

# Plotting the bar graph for weekdays and weekends
labels = ['Weekday', 'Weekend']
values = [weekday_avg, weekend_avg]

plt.bar(labels, values, color=['blue', 'green'])

# Adding titles and labels
plt.title('Average Power Consumption on Weekdays vs Weekends')
plt.ylabel('Average Global Active Power (kWh)')
plt.tight_layout()
plt.show()
```

## 4. Find days where energy consumption exceeded 5 kWh.

```python
daily_energy = df['Global_active_power'].resample('D').sum()
high_usage_days = daily_energy[daily_energy > 5]
print(high_usage_days)
```

```python
# Plotting the high usage days
plt.figure(figsize=(10, 6))
plt.plot(high_usage_days.index, high_usage_days, marker='o', linestyle='-', color='r', label='High Power Usage Days')
plt.title('High Power Usage Days (Greater than 5 kWh)')
plt.xlabel('Date')
plt.ylabel('Global Active Power (kWh)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 5. Plot the energy usage trend for the first week of January 2007.

```python
jan_week = df['2007-01-01':'2007-01-07']
jan_week['Global_active_power'].plot(figsize=(12, 4), title='Energy Usage - Jan 1 to Jan 7, 2007')
plt.ylabel('Global Active Power (kW)')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```

## 	6. Find the average voltage for each day of the first week of February 2007.

```python
feb_week = df['2007-02-01':'2007-02-07']
daily_voltage_avg = feb_week['Voltage'].resample('D').mean()
print(daily_voltage_avg)
```

```python
# Plotting the daily voltage average
plt.figure(figsize=(10, 6))
plt.plot(daily_voltage_avg.index, daily_voltage_avg, marker='o', linestyle='-', color='purple', label='Daily Voltage Average')
plt.title('Daily Average Voltage (First Week of February 2007)')
plt.xlabel('Date')
plt.ylabel('Average Voltage (V)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 7. What is the correlation between global active power and sub-metering values?

```python
cols = ['Global_active_power', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
df[cols] = df[cols].astype(float)
correlation = df[cols].corr()
print(correlation)
```

```python
# Plotting the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Power Consumption and Sub-Metering Values')
plt.tight_layout()
plt.show()
```
