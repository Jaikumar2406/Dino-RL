```python
from mss import mss
import pydirectinput
import pytesseract
import time
import matplotlib.pyplot as plt
import gymnasium as gym
from gymnasium import Env
from gymnasium.spaces import Box, Discrete
import cv2
import numpy as np
```

```python
class dinoai(Env):
    def __init__(self):
        super().__init__()
        self.observation_space = Box(low =0 , high = 255 , shape = (1,83,100) , dtype= np.uint8)
        self.action_space = Discrete(3)

        self.cap = mss()
        self.game_location = {"top":300 , "left":0, "width":600 , "height":300}
        self.game_done = {"top":250 , "left":630, "width": 660, "height":70}
        

    def step(self , action):
        action_map={
            0:'space',
            1:'down',
            2:'no_op'
        }
        if action_map !=2:
            pydirectinput.press(action_map[action])

        done , done_cap = self.get_close()
        observation = self.get_observation()
        reward = 1
        terminated = done
        truncated = False
        info = {}
        return observation, reward, terminated, truncated, info

        
    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        time.sleep(1)
        pydirectinput.click(x =150 , y=150)
        pydirectinput.press("space")
        observation = self.get_observation()
        info = {}
        return observation, info

        
    def render(self):
        cv2.imshow("GAME" , np.array(self.cap.grab(self.game_location))[:,:,:3])
        if cv2.waitKey(10) & 0xff == ord("q"):
            self.close

    def close(self):
        cv2.destroyAllWindows()

    def get_observation(self):
        raw = np.array(self.cap.grab(self.game_location))[:,:,:3].astype(np.uint8)#screen capture
        gray = cv2.cvtColor(raw , cv2.COLOR_BGR2GRAY)#convert image into gray
        resized = cv2.resize(gray , (100,83))#resize the image
        channel = np.reshape(resized , (1 , 83 ,100))# first channel
        return channel
    def get_close(self):
        done_cap = np.array(self.cap.grab(self.game_done))[:, :, :3]
        gray = cv2.cvtColor(done_cap, cv2.COLOR_BGR2GRAY)
        done_strings = ["GAME", "GAHE", "GANE"]
        # done = False
        res = pytesseract.image_to_string(gray).strip().upper()[:4]
        done = any(word in res for word in done_strings)
        return done , done_cap
```

```python
class TrainAndLoggingCallback(BaseCallback):

    def __init__(self, check_freq, save_path, verbose=1):
        super(TrainAndLoggingCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.save_path = save_path

    def _init_callback(self):
        if self.save_path is not None:
            os.makedirs(self.save_path, exist_ok=True)

    def _on_step(self):
        if self.n_calls % self.check_freq == 0:
            model_path = os.path.join(self.save_path, 'best_model_{}'.format(self.n_calls))
            self.model.save(model_path)

        return True
```

```python
CHECKPOINT_DIR = './train/'
LOG_DIR = './logs/'
```

```python
callback = TrainAndLoggingCallback(check_freq=1000, save_path=CHECKPOINT_DIR)
```

```python
model.learn(total_timesteps=100000, callback=callback)
```
